```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def before_wrtie_csv(df):
        df = df.T
        new_header = df.iloc[0]
        df = df[1:]
        df.columns = new_header
        df=df.iloc[:,[i%2==0 for i in range(len(df.columns))]]
        return df
    
def plot_test_acc(title):
    if title == 'Test Accuracy with Different Learning Rate':
        df = test_lr = pd.read_csv('output_lr_test.csv',sep = ':',header = None)
        df_new = before_wrtie_csv(df)
        df_new.to_csv('test_lr.csv', header=False)
        df = pd.read_csv('test_lr.csv',header = None)
    elif title == 'Test Accuracy with Different Number of Iterations with Learning Rate Equals to 2':
        df = pd.read_csv('output_110_test.csv',sep = ':',header = None)
        df_new = before_wrtie_csv(df)
        df_new.to_csv('test_iter.csv', header=False)
        df = pd.read_csv('test_iter.csv',header = None)
    df = df.T
    df.iloc[0] = [0]
    
    plt.figure(figsize=(10,8))
    plt.plot(df)
    plt.ylabel('Test Accuracy')
    plt.title(title)
    if title == 'Test Accuracy with Different Learning Rate':
        plt.xlabel('Learning Rate')
        plt.savefig('test_acc_lr')
    elif title == 'Test Accuracy with Different Number of Iterations with Learning Rate Equals to 2':
        plt.xlabel('Number of Iterations')
        plt.savefig('test_acc_iter')
    plt.show()

def plot_train_acc(title):
    if title == 'Train Accuracy with Different Learning Rates':
        df_dev = pd.read_csv('output_lr_dev.csv',sep = ':',header = None)
        df_dev_new = before_wrtie_csv(df_dev)
        df_dev_new.to_csv('dev_lr.csv', header=False)
        df_dev = pd.read_csv('dev_lr.csv',header = None)
        df_train = pd.read_csv('output_lr_train.csv',sep = ':',header = None)
        df_train_new = before_wrtie_csv(df_train)
        df_train_new.to_csv('train_lr.csv', header=False)
        df_train = pd.read_csv('train_lr.csv',header = None)
    elif title == 'Train Accuracy with Different Number of Iterations with Learning Rate Equals to 2':
        df_dev = pd.read_csv('output_110_dev.csv',sep = ':',header = None)
        df_dev_new = before_wrtie_csv(df_dev)
        df_dev_new.to_csv('dev_iter.csv', header=False)
        df_dev = pd.read_csv('dev_iter.csv',header = None)
        df_train = pd.read_csv('output_110_train.csv',sep = ':',header = None)
        df_train_new = before_wrtie_csv(df_train)
        df_train_new.to_csv('train_iter.csv', header=False)
        df_train = pd.read_csv('train_iter.csv',header = None)
    df_dev = df_dev.T
    df_dev.iloc[0] = [0]
    df_train = df_train.T
    df_train.iloc[0] = [0]
    
    plt.figure(figsize=(10,8))
    plt.plot(df_train, marker = '', color = 'red', linewidth = 2,label = 'Training Set')
    plt.plot(df_dev, marker = '', color = 'blue', linewidth = 1, label = 'Development Set')
    plt.ylabel('Train Accuracy')
    plt.title(title)
    plt.legend()
    if title == 'Train Accuracy with Different Learning Rates':
        plt.xlabel('Learning Rate')
        plt.savefig('train_dev_acc_lr')
    elif title == 'Train Accuracy with Different Number of Iterations with Learning Rate Equals to 2':
        plt.xlabel('Number of Iterations')
        plt.savefig('train_dev_acc_iter')
    plt.show()
```

```python
#Test Accuracy Plot
title_test = ['Test Accuracy with Different Learning Rate', 'Test Accuracy with Different Number of Iterations with Learning Rate Equals to 2']
for title in title_test:
    plot_test_acc(title)
```

```python
# Train Accuracy Plot
title_train = ['Train Accuracy with Different Learning Rates', 'Train Accuracy with Different Number of Iterations with Learning Rate Equals to 2']
for title in title_train:
    plot_train_acc(title)
```
