```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def before_wrtie_csv(df):
        df = df.T
        new_header = df.iloc[0]
        df = df[1:]
        df.columns = new_header
        return df
```

```python
def plot_test(title):
    if title == 'Test Accuracy with Different Hidden Dim':
        df = pd.read_csv('output_hd_test.csv', sep = ':', header = None)
        df_new = df_new = before_wrtie_csv(df)
        df_new.to_csv('test_hd.csv', header=False)
        df = pd.read_csv('test_hd.csv', header= None)
    elif title == 'Test Accuracy with Different Learning Rate':
        df = pd.read_csv('output_lr_test.csv', sep = ':', header = None)
        df_new = before_wrtie_csv(df)
        df_new.to_csv('test_lr.csv', header=False)
        df = pd.read_csv('test_lr.csv', header= None)
    elif title == 'Test Accuracy with Different Number of Iterations with Learning Rate Equals to 0.01':
        df = pd.read_csv('output_iter_test.csv', sep = ':', header = None)
        df_new = df_new = before_wrtie_csv(df)
        df_new.to_csv('test_iter.csv', header=False)
        df = pd.read_csv('test_iter.csv', header= None)
    df = df.T
    df.iloc[0] = [0]
    plt.figure(figsize=(10,8))
    plt.plot(df)
    plt.ylabel('Test Accuracy')
    plt.title(title)
    if title == 'Test Accuracy with Different Hidden Dim':
        plt.xlabel('Hidden Dim')
        plt.savefig('test_hd')
    elif title == 'Test Accuracy with Different Learning Rate':
        plt.xlabel('Learning Rate')
        plt.savefig('test_lr')
    elif title == 'Test Accuracy with Different Number of Iterations with Learning Rate Equals to 0.01':
        plt.xlabel('Number of Iterations')
        plt.savefig('test_iter')
    plt.show()
```

```python
plot_test('Test Accuracy with Different Hidden Dim')
plot_test('Test Accuracy with Different Learning Rate')
plot_test('Test Accuracy with Different Number of Iterations with Learning Rate Equals to 0.01')
```

```python
def plot_train_dev(title):
    if title == 'Train Accuracy with Different Hidden Dim':
        df_dev = pd.read_csv('output_hd_dev.csv',sep = ':',header = None)
        df_dev_new = before_wrtie_csv(df_dev)
        df_dev_new.to_csv('dev_hd.csv', header=False)
        df_dev = pd.read_csv('dev_hd.csv',header = None)
        df_train = pd.read_csv('output_hd_train.csv',sep = ':',header = None)
        df_train_new = before_wrtie_csv(df_train)
        df_train_new.to_csv('train_hd.csv', header=False)
        df_train = pd.read_csv('train_hd.csv',header = None)
    elif title == 'Train Accuracy with Different Learning Rate':
        df_dev = pd.read_csv('output_lr_dev.csv',sep = ':',header = None)
        df_dev_new = before_wrtie_csv(df_dev)
        df_dev_new.to_csv('dev_lr.csv', header=False)
        df_dev = pd.read_csv('dev_lr.csv',header = None)
        df_train = pd.read_csv('output_lr_train.csv',sep = ':',header = None)
        df_train_new = before_wrtie_csv(df_train)
        df_train_new.to_csv('train_lr.csv', header=False)
        df_train = pd.read_csv('train_lr.csv',header = None)
    elif title == 'Train Accuracy with Different Number of Iterations with Learning Rate Equals to 0.01':
        df_dev = pd.read_csv('output_iter_dev.csv',sep = ':',header = None)
        df_dev_new = before_wrtie_csv(df_dev)
        df_dev_new.to_csv('dev_iter.csv', header=False)
        df_dev = pd.read_csv('dev_iter.csv',header = None)
        df_train = pd.read_csv('output_iter_train.csv',sep = ':',header = None)
        df_train_new = before_wrtie_csv(df_train)
        df_train_new.to_csv('train_iter.csv', header=False)
        df_train = pd.read_csv('train_iter.csv',header = None)
    df_dev = df_dev.T
    df_dev.iloc[0] = [0]
    df_train = df_train.T
    df_train.iloc[0] = [0]
    plt.figure(figsize=(10,8))
    plt.plot(df_train, marker = '', color = 'red', linewidth = 2,label = 'Training Set')
    plt.plot(df_dev, marker = '', color = 'blue', linewidth = 1, label = 'Development Set')
    plt.ylabel('Train Accuracy')
    plt.title(title)
    plt.legend()
    if title == 'Train Accuracy with Different Hidden Dim':
        plt.xlabel('Hidden Dim')
        plt.savefig('train_dev_acc_hd')
    elif title == 'Train Accuracy with Different Learning Rates':
        plt.xlabel('Learning Rate')
        plt.savefig('train_dev_acc_lr')
    elif title == 'Train Accuracy with Different Number of Iterations with Learning Rate Equals to 2':
        plt.xlabel('Number of Iterations')
        plt.savefig('train_dev_acc_iter')
    plt.show()
```

```python
plot_train_dev('Train Accuracy with Different Hidden Dim')
plot_train_dev('Train Accuracy with Different Learning Rate')
plot_train_dev('Train Accuracy with Different Number of Iterations with Learning Rate Equals to 0.01')
```
